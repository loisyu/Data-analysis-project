# DATA CLEANING & TRANSFORMATION
## *DS JOBS ON GLASSDOR*

```python
import numpy as np
import pandas as pd
```

```python
job_data = pd.read_csv ("/kaggle/input/data-science-job-posting-on-glassdoor/Uncleaned_DS_jobs.csv")
```

```python
job_data.head()
```

```python
job_data.dropna()
col_names = list(job_data.columns) #list of column names to see which colunm is irrelevant to use 
```

```python
col_names
```

```python
job_data= job_data.drop(['index', 'Size', 'Founded', 'Type of ownership', 'Competitors', 'Revenue', 'Headquarters'], axis=1)
```

```python
job_data.head()
```

```python
# remove the number in company name column
job_data['Company Name'] = job_data['Company Name'].str.replace('\d+', '', regex=True)
```

```python
job_data.head()
```

```python
#remove \n in the company name column
job_data['Company Name'] = job_data['Company Name'].str.replace('\n', '')
job_data.head()
```

```python
#remove (Glassdoor est.) in the salart estimate column
job_data['Salary Estimate'] = job_data['Salary Estimate'].str.replace(r"\(.*\)","", regex = True)
job_data.head()
```

```python
#replace K in salary into 000
job_data['Salary Estimate'] = job_data['Salary Estimate'].str.replace('K', '000')
job_data.head()
```

```python
#remove unwanted words in job description
job_data['Job Description'] = job_data['Job Description'].str.replace('\n', '')
job_data.head()
```

```python
#split location column into City & State
job_data[['City', 'State']] = job_data['Location'].str.split(', ', n=1, expand=True)
job_data.head()
```

```python
#drop location column
job_data= job_data.drop('Location', axis=1)
job_data.head()
```

```python
job_data = job_data.reset_index(drop=True)
job_data.head()
```

```python
job_data.dtypes
```

```python
# Function to filter dataset for a specific city
def city_filter(city):
    filtered_city = job_data[job_data['City'] == city]
    return filtered_city
```

```python
city_filter('New York')
```

```python
#Function to filter dataset for a specific state
def state_filter(state):
    filtered_state = job_data[job_data['State'] == state]
    return filtered_state
```

```python
state_filter('MA')
```

```python
# filter company based on a minimum rating
def rating_filter(rating):
    filtered_rating = job_data[job_data['Rating'] >= rating]
    return filtered_rating
```

```python
rating_filter(4.0)
```
